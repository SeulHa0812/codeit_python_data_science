```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
year_array = np.array([2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])

stock_array = np.array([
    14.46, 19.01, 20.04, 27.59, 26.32,
    28.96, 42.31, 39.44, 73.41, 132.69
])
```

```python
# line graph
plt.plot(year_array, stock_array) # (x axis, y axis)
plt.show() # 생략 가능
```

```python
name_array = np.array(['mark', 'dongwook', 'hyojune', 'sowon', 'taeho'])
votes_array = np.array([5, 10, 6, 8, 3])
```

```python
# bar graph
plt.bar(name_array, votes_array)
plt.show()
```

```python
height_array = np.array([
    165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
    165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
    172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

weight_array = np.array([
    62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
    57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
    67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])
```

```python
# scatter plot ( 두 변수의 상관관계 파악 )
plt.scatter(height_array, weight_array)
plt.show()
```

```python
# plt.rcParams['figure.figsize'] = (5, 5) 일괄적 크기 변경
plt.figure(figsize=(10, 4)) 
plt.scatter(height_array, weight_array, c='red', marker='s') # 세번째 파라미터 c = color, hex code 가능
# 네번째 파라미터 점 모양 s = square 약자 https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
plt.title('Height and Weight') # title
plt.xlabel('Height (cm)') # x label
plt.ylabel('Weight (kg)') # y label
plt.show()
```
