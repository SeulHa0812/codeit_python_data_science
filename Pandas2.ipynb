```python
# CSV = Comma-Separated-Values
# 첫 줄에 column명이 기재되어 있는지 헤더(Header)확인 후 사용
import pandas as pd
```

```python
burger_df = pd.read_csv("data/burger.csv", index_col="product_name") # 헤더 자동 인식 
# 헤더가 없다면 두번째 파라미터로 header=None
# column명 설정법 파라미터로 names=["product_name","calories","carb","protein","fat","sodium","category"]
burger_df
```

```python
# iloc(integer location), loc(컬럼 이름 or 로우 이름 통해)
burger_df.iloc[3, 1]
```

```python
burger_df.iloc[[2, 3], [1, 2]]
```

```python
burger_df.iloc[1:4, :3]
```

```python
burger_df.loc['Double Whopper', 'protein']
```

```python
burger_df.loc[['Whopper', 'Double Whopper'], 'protein']
```

```python
burger_df.loc[['Whopper', 'Double Whopper'], ['carb', 'protein', 'fat']]
```

```python
burger_df.loc['Double Whopper':'Bacon King', 'calories':'fat'] # 마지막 값 포함
```
